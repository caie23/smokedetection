```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8.yaml')  # build a model from scratch

# Train the model
model.train(task='detect', mode='train', data='smokedata/smoke.yaml', epochs=100, batch=16, name='v8_100e_16b')

# use the mode lto predict test data
results = model(source='smokedata/images/test/', save=True, save_txt=True)

# export model as onnx format
model.export(format='onnx')
```
